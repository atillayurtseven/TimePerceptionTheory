```python
import numpy as np
import matplotlib.pyplot as plt

def calc_age_percention(young_person_age, old_person_age, years_after=20):
    # Generate an array of years from 1 to the specified years_after
    years = np.arange(1, years_after + 1)
    
    # Initial ages of the young and old individuals
    young = young_person_age
    old = old_person_age
    
    # Lists to store time perception values for the young and old individuals
    yp = []
    op = []
    
    # Calculate time perception for each year
    for i in range(len(years)):
        # Calculate time perception for the young individual
        c = (young + 1) / young - 1
        # Calculate time perception for the old individual
        o = (old + 1) / old - 1
        # Append calculated values to the lists
        yp.append(c * 100)
        op.append(o * 100)
        # Increment ages for the next year
        young += 1
        old += 1
    
    # Create a figure and axis for the plot
    fig, ax = plt.subplots(1, figsize=(10, 8))
    
    # Create custom x-axis ticks and add vertical lines for each tick
    custom_xticks = []
    for i in range(2, years_after + 2, 2):
        ax.axvline(i, ls="-.", color="black", lw="0.15")
        custom_xticks.append(i)
    
    # Add horizontal lines for each time perception value
    for i in yp:
        ax.axhline(i, ls="-.", color="black", lw="0.15")
    for i in op:
        ax.axhline(i, ls="-.", color="black", lw="0.15")
    
    # Highlight the initial time perception for the old individual with a red line
    plt.axhline(op[0], ls="-.", color="red", lw="0.75")
    
    # Set custom x-axis ticks and labels
    ax.set_xticks(custom_xticks)
    ax.set_xticklabels(custom_xticks)
    
    # Set x and y axis labels and plot title
    plt.xlabel("Years")
    plt.ylabel("Time Perception (%)")
    plt.title("Theory of Age-Dependent Time Perception Change by Atilla Yurtseven")
    
    # Labels for the plot legend
    ylabel = f"{young_person_age} Years Old Time Perception"
    olabel = f"{old_person_age} Years Old Time Perception"
    
    # Plot time perception values for both young and old individuals
    ax.plot(years, yp, marker="o", label=ylabel)
    ax.plot(years, op, marker="o", label=olabel)
    
    # Display the legend and the plot
    plt.legend()
    plt.show()

# Calculate and visualize time perception for young (50 years old) and old (70 years old) individuals over 21 years
calc_age_percention(50, 70, 21)
```
